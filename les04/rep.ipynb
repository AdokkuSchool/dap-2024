```python
import numpy as np
import pandas as pd

url_titanic = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url_titanic)
df.set_index('PassengerId', inplace=True)

print("Данные успешно загружены!")
```

```python
print("Задание 3 (Описание данных):")
print(df.describe())
print("-" * 30)
```

```python
print("Задание 4 (Количество мужчин и женщин):")
print(df['Sex'].value_counts())
print("-" * 30)
```

```python
print("Задание 5:")
print("Распределение по классам (Всего):\n", df['Pclass'].value_counts())

print("\nРаспределение по классам (по полу):")
print(df.groupby(['Sex', 'Pclass']).size())

male_2nd_class = df[(df['Sex'] == 'male') & (df['Pclass'] == 2)].shape[0]
print(f"\nМужчин во 2-м классе: {male_2nd_class}")
print("-" * 30)
```

```python
fare_median = round(df['Fare'].median(), 2)
fare_std = round(df['Fare'].std(), 2)

print("Задание 6:")
print(f"Медиана Fare = {fare_median}")
print(f"СКО Fare = {fare_std}")
print("-" * 30)
```

```python
young_survived = df[df['Age'] < 30]['Survived'].mean()
old_survived = df[df['Age'] > 60]['Survived'].mean()

print("Задание 7:")
print(f"Доля выживших (<30 лет): {young_survived:.2%}")
print(f"Доля выживших (>60 лет): {old_survived:.2%}")
print(f"Молодые выживали чаще? {young_survived > old_survived}")
print("-" * 30)
```

```python
female_survived = df[df['Sex'] == 'female']['Survived'].mean()
male_survived = df[df['Sex'] == 'male']['Survived'].mean()

print("Задание 8:")
print(f"Доля выживших женщин: {female_survived:.2%}")
print(f"Доля выживших мужчин: {male_survived:.2%}")
print(f"Женщины выживали чаще? {female_survived > male_survived}")
print("-" * 30)
```

```python
def get_first_name(name):
    try:
        return name.split(',')[1].split('.')[1].strip().split(' ')[0]
    except IndexError:
        return name.split(' ')[0]

male_names = df[df['Sex'] == 'male']['Name'].apply(get_first_name)
top_name = male_names.value_counts().idxmax()

print(f"Задание 9: Самое популярное мужское имя: {top_name}")
print("-" * 30)
```

```python
print("Задание 10:")
age_stats = df.groupby(['Pclass', 'Sex'])['Age'].mean()
print(age_stats)

print("\nПроверка утверждений:")
print(f"1. Мужчины 1-го класса > 40? {age_stats[1, 'male'] > 40}")
print(f"2. Женщины 1-го класса > 40? {age_stats[1, 'female'] > 40}")

check_3 = all(age_stats[i, 'male'] > age_stats[i, 'female'] for i in [1, 2, 3])
print(f"3. Мужчины всех классов старше женщин? {check_3}")

mean_ages = df.groupby('Pclass')['Age'].mean()
check_4 = (mean_ages[1] > mean_ages[2]) and (mean_ages[2] > mean_ages[3])
print(f"4. В 1 классе старше, чем во 2, а во 2 старше 3? {check_4}")
print("-" * 30)
```

## Дополнительные вопросы для защиты


Определения
Медиана: Это число, которое находится ровно посередине упорядоченного (от меньшего к большему) набора чисел. Половина чисел меньше медианы, половина — больше.

Пример: Для ряда 1, 3, 10 медиана — 3. Для ряда 1, 3, 5, 10 медиана — 4 (среднее между 3 и 5).

Дисперсия: Мера разброса данных. Показывает, насколько сильно числа отклоняются от среднего значения. Рассчитывается как средний квадрат отклонений.

Среднеквадратичное отклонение (СКО / std): Это корень из дисперсии. Оно показывает разброс в тех же единицах измерения, что и сами данные (например, в годах или долларах).

Series vs DataFrame
Series: Это одномерный массив данных (как один столбец в Excel), у которого есть метки (индекс).

DataFrame: Это двумерная таблица данных (как весь лист в Excel), состоящая из строк и столбцов. По сути, это набор Series, объединенных общим индексом.

Изменение индекса в Series
Да, можно. Можно присвоить новый список индексов через атрибут .index:

Python s.index = ['new1', 'new2', ...] 4. Сложение двух Series

Операции в Pandas всегда выравниваются по индексу (по меткам), а не по порядку расположения.

Одинаковые типы, размеры и индексы: Сложение происходит поэлементно (значение с индексом 'A' сложится со значением с индексом 'A').

Одинаковые типы, размеры, но РАЗНЫЕ индексы: Результатом будет объединение индексов. Там, где индексы совпали — сумма. Там, где не совпали — NaN (пустое значение).

Разные размеры: Аналогично, выравнивание идет по индексу. Недостающие индексы дадут NaN.

Разные типы данных:

Число + Строка = Ошибка.

Число + Число (int + float) = Float.

iloc vs loc
loc: Выбор данных по МЕТКЕ индекса (названию).

Пример: df.loc['Ivanov'] (найдет строку, где индекс называется Ivanov).

Включает последнюю границу при срезе.

iloc: Выбор данных по ПОЗИЦИИ (порядковому номеру, как в обычном списке).

Пример: df.iloc[0] (найдет самую первую строку).

НЕ включает последнюю границу при срезе.

Добавление Series в DataFrame
Да, можно. Это основной способ добавления нового столбца.

Python df['NewColumn'] = my_series Pandas автоматически сопоставит данные по индексам.
