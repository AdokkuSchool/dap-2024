```python
class Solution1(object):
    def twoSum(self, nums, target):
        seen = {}
        for i, num in enumerate(nums):
            complement = target - num
            if complement in seen:
                return [seen[complement], i]
            seen[num] = i
        return []


class Solution2(object):
    def isPalindrome(self, x):
        if x < 0:
            return False
        return str(x) == str(x)[::-1]


class Solution3(object):
    def romanToInt(self, s):
        roman_map = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        total = 0
        prev_value = 0
        for char in reversed(s):
            value = roman_map[char]
            if value < prev_value:
                total -= value
            else:
                total += value
            prev_value = value
        return total


class Solution4(object):
    def longestCommonPrefix(self, strs):
        if not strs:
            return ""
        strs.sort()
        s1 = strs[0]
        s2 = strs[-1]
        i = 0
        while i < len(s1) and i < len(s2) and s1[i] == s2[i]:
            i += 1
        return s1[:i]


class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution5(object):
    def mergeTwoLists(self, list1, list2):
        dummy = ListNode()
        current = dummy
        while list1 and list2:
            if list1.val < list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next
        if list1:
            current.next = list1
        elif list2:
            current.next = list2
        return dummy.next


class Solution(object):
    def lengthOfLongestSubstring(self, s):
        char_map = {}
        left = 0
        max_len = 0
        for right, char in enumerate(s):
            if char in char_map and char_map[char] >= left:
                left = char_map[char] + 1
            char_map[char] = right
            max_len = max(max_len, right - left + 1)
        return max_len
```

```python
print("--- ЗАПУСК ПРОВЕРКИ ---")

sol1 = Solution1()
print("Задача 1:", sol1.twoSum([2,7,11,15], 9))

sol2 = Solution2()
print("Задача 2:", sol2.isPalindrome(121))

sol3 = Solution3()
print("Задача 3:", sol3.romanToInt("MCMXCIV"))

sol4 = Solution4()
print("Задача 4:", sol4.longestCommonPrefix(["flower","flow","flight"]))

l1 = ListNode(1, ListNode(2, ListNode(4)))
l2 = ListNode(1, ListNode(3, ListNode(4)))
sol5 = Solution5()
res = sol5.mergeTwoLists(l1, l2)
out = []
while res:
    out.append(res.val)
    res = res.next
print("Задача 5:", out)

sol6 = Solution()
print("Задача 6:", sol6.lengthOfLongestSubstring("abcabcbb"))
```

## Ответы на контрольные вопросы


1. Что такое динамическая типизация и чем она отличается от статической?
Динамическая типизация (Python): Тип переменной определяется автоматически в момент присваивания значения (во время выполнения программы). Переменная может менять свой тип по ходу кода (сначала число, потом строка). Статическая типизация (C++, Java): Тип переменной задается строго при её объявлении (во время компиляции) и измениться не может.

2. Чем отличается цикл for в Python от цикла for в С++?
В Python: Цикл for работает как итератор (аналог foreach). Он перебирает элементы коллекции (списка, строки) по очереди.
for item in [1, 2, 3]: # берет элементы
В C++: Классический for управляется счетчиком, условием и шагом изменения.
for (int i=0; i<3; i++) // работает с индексами
3. Для чего предназначены операторы continue и break?
break: Немедленно прерывает выполнение цикла и выходит из него.
continue: Пропускает оставшуюся часть текущей итерации и переходит к следующему витку цикла.

5. Что такое лямбда функции и для чего они предназначены?

Lambda-функция — это анонимная функция, которая записывается в одну строку. Используется там, где нужна короткая функция для одноразового использования (например, внутри sort, map, filter), чтобы не объявлять полноценную функцию через def. Пример: lambda x: x * 2

6. Может ли тело функции содержать определение другой функции?

Да, может. Это называется вложенными функциями. Внутренняя функция видна только внутри внешней. Используется для создания замыканий (closures) и декораторов.

7. Как определить класс в python? Для чего нужна переменная self?

Класс определяется словом class ИмяКласса:. Переменная self — это ссылка на текущий экземпляр (объект) класса. Она обязательна первым аргументом в методах, чтобы метод понимал, с данными какого именно объекта он работает (аналог this в других языках, но явный).

8. Как определить приватный метод? Можно ли вызвать приватный метод за пределами класса?

Приватный метод начинается с двух нижних подчеркиваний: def __method(self):. Напрямую вызвать его нельзя (obj.__method() выдаст ошибку). Однако Python использует "искажение имен" (name mangling), поэтому технически вызвать его можно так: _ClassName__method(), но это крайне не рекомендуется.

9. Как определить статический метод класса? Для чего может понадобиться статический метод?

Используется декоратор @staticmethod. Такой метод не принимает self и не зависит от состояния конкретного объекта. Он нужен для функций, которые логически относятся к классу (например, сервисные функции), но не используют данные объекта.

10. Чем staticmethod отличается от classmethod?

@staticmethod: Обычная функция внутри класса. Не знает ни про объект (self), ни про класс.

@classmethod: Принимает первым аргументом ссылку на класс (cls), а не на объект. Часто используется для создания альтернативных конструкторов (Factory methods).

12. Что такое кортеж, для чего он используется?

Кортеж (tuple) — это неизменяемый список. Записывается в круглых скобках (1, 2, 3). Используется для:

Защиты данных от случайного изменения.

Использования в качестве ключей словаря (списки нельзя, кортежи можно).

Экономии памяти (работает быстрее списка).
